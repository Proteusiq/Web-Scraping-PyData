## Discovering APIs
How to follow your browsers behavior using TrustPilot as example

```python
import json
from requests_html import HTMLSession
```

```python
s = HTMLSession()
```

```python
headers = {
'referer': 'https://www.trustpilot.com/',
'user-agent': ('Mozilla/5.0 (X11; Linux x86_64) '
               'AppleWebKit/537.36 (KHTML, like Gecko) '
               'Chrome/75.0.3770.100 Safari/537.36')
}
```

```python
s.headers.update(headers)
```

```python
QUERY_URL = 'https://www.trustpilot.com/businessunit/search'
```

```python
params = {
    'query': 'www.skat.dk',
    'country': 'DK',
    'perpage': 1
}
```

```python
res = s.get(QUERY_URL, params=params)
```

```python
res.json()
```

```python
query = res.json()
business_id = query['businessUnits'][0]['id']
business_name = query['businessUnits'][0]['displayName']
```

```python
business_id
```

```python
f'https://www.trustpilot.com/domains/ratingdistributionchart/{business_id}'
```

```python
# Two Path
# Less Ninja HTML(Weird 1) & No Ninja HTML

call_params = {
    'languages': 'all',
    'page': 1
}

r = s.get(f'https://www.trustpilot.com/review/{params.get("query")}',params=call_params)
```

```python
r.url
```

```python
# Find out what script are gathering as data
for index, url in enumerate(r.html.find('script')):
    print(index, url.text)
```
