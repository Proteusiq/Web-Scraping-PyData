```python
import pandas as pd
import requests

r = requests.get('https://sslproxies.org/')
proxies_df = pd.read_html(r.text, attrs={'id':'proxylisttable'}, flavor='lxml', skiprows=[101])[0]
proxies_df
```

```python
index = 2
print(proxies_df.loc[index,])

print('\nProxy selected:')
print(f"\t{proxies_df.loc[index,'IP Address']}:{proxies_df.loc[index, 'Port']}")
```
