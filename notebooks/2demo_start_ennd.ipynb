```python
from collections import defaultdict

from bs4 import BeautifulSoup
import pandas as pd
from requests import Session
```

```python
session = Session()
```

```python
session.headers = {'User-Agent': 
                   'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36'\
                   '(KHTML, like Gecko) Chrome/75.0.3770.80 Safari/537.36'}
```

```python
URL = 'https://www.autouncle.dk/en/cars_search'
```

```python
store = defaultdict(list)
```

```python
def get_cars(page_nr=1):
    
    stop = False
    while not stop:
        print(f'Get Data from page {page_nr}')
        params = {'page':page_nr}
        r = session.get(URL, params=params)
        soup = BeautifulSoup(r.content,'lxml')
        items = soup.find_all('div',class_='listing-item')
        for item in items:
            store['car'].append(item.find('img')['alt'])
            store['cost'].append(item.find('span', class_='price')['content'])
            store['should_cost'].append(item.find('span', class_='should-cost').find_all('span')[-1].get_text(strip=True))
            [store[i['class'][-1]].append(i.span.get_text(strip=True))  for i in item.find_all('li')]
    
        # next is disabled when no more page
        stop = 'disabled' in soup.find('div',{'id':'pagination'}).find_all('span')[-1]['class']
        page_nr += 1
    print('Data Collection Completed')
    return store
     
    
```

```python
_ = get_cars()
```

```python
df = pd.DataFrame(store)
```

```python
df.to_pickle('car_data.pkl')
```
