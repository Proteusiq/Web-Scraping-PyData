## JUST PANDAS

```python
import pandas as pd
from requests import Session

s = Session()
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) '\
                         'AppleWebKit/537.36 (KHTML, like Gecko) '\
                         'Chrome/75.0.3770.80 Safari/537.36',
          'Content-Type': 'application/json;charset=UTF-8'}
# Add headers
s.headers.update(headers)
```

```python
WIKI_URL = 'https://en.wikipedia.org/wiki/List_of_most_expensive_films'

w = res = s.get(WIKI_URL)
```

```python
pd.read_html(w.content)[0].head()
```

# Better Way is Network

```python
BASE_URL = 'https://footballapi.pulselive.com/football/standings'


headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) '\
                         'AppleWebKit/537.36 (KHTML, like Gecko) '\
                         'Chrome/75.0.3770.80 Safari/537.36',
          'Content-Type': 'application/json;charset=UTF-8',
          'Origin': 'https://www.premierleague.com',
          'Referer': 'https://www.premierleague.com/tables?co=1&se=210&ha=-1'}

s.headers.update(headers)

params = {'compSeasons': 210,
    'altIds': True,
    'detail': 2,
    'FOOTBALL_COMPETITION': 1}
```

```python
r = s.get(BASE_URL, params=params)
```

```python
r.ok
```

```python
r.json()
```
